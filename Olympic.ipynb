### Find the number of each medal for each year.
##### The output should look like this 
Capture.PNG

```python
new_df = df[['Year', 'Medal']].dropna()
n_df = new_df.groupby('Year')['Medal'].value_counts().reset_index("Year")
n_df.columns = ['Year','Count']
n_df.reset_index(inplace=True)
piv = n_df.pivot(index="Year", columns="Medal", values="Count")
final_df = piv[['Gold','Silver','Bronze']]
final_df.rename_axis(columns=None)
final_df.columns = ['N_Gold', 'N_Silver', 'N_Bronze']
final_df
# Output doesn't match. Please share your own way in comments. Thanks!
```
